```python
# import dependencies
import pandas as pd
import json
```

```python
df = pd.read_csv("data_files/nba_data.csv")
df.head()
```

```python
df = df.loc[df['season']==2023]
```

```python
df_visitor = df[['season','visitor_name','visitor_points_x','visitor_assists','visitor_totReb','visitor_steals','visitor_fgm','visitor_tpm','visitor_turnovers','visitor_blocks']]
```

```python
df_home = df[['season','home_name','home_points_x','home_assists','home_totReb','home_steals','home_fgm','home_tpm','home_turnovers','home_blocks']]
```

```python
df_visitor = df_visitor.rename(columns={'visitor_name':'team_name','visitor_points_x':'points','visitor_assists':'assists','visitor_totReb':'rebounds','visitor_steals':'steals','visitor_fgm':'fgm','visitor_tpm':'3ptr','visitor_turnovers':'turnovers','visitor_blocks':'blocks'})
df_visitor.head()
```

```python
df_home = df_home.rename(columns={'home_name':'team_name','home_points_x':'points','home_assists':'assists','home_totReb':'rebounds','home_steals':'steals','home_fgm':'fgm','home_tpm':'3ptr','home_turnovers':'turnovers','home_blocks':'blocks'})
df_home.head()
```

```python
df_all = pd.concat([df_home,df_visitor], axis=0)
df_all
```

```python
df_group = df_all.groupby('team_name')[['points','assists','rebounds','steals','fgm','3ptr','turnovers','blocks']].mean()
```

```python
df_group
```

```python
df_group = df_group.reset_index()
```

```python
df_group.columns
```

```python
df_group = df_group[~df_group['team_name'].str.contains("flamengo|Ra'anana Maccabi Ra'anana", case=False, na=False)]
df_group
```

```python
df_dict = df_group.to_dict(orient='records')
```

```python
df_dict
```

```python
# team stats dictionary
stats_json = json.dumps(df_dict, indent=4)
json.dump(df_dict, open('team_stats_dict.json', 'w'), indent=4)
```
